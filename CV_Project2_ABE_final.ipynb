```python
from ultralytics import YOLO
import cv2
from datetime import datetime

model = YOLO("Backery_best_ABE.pt")

# Program Functions 
    #1. picture detection and checkup
#results = model.track("Test3.jpg", show=True)
    #2. Realtime webcam detection (only detection no checkup cus its not reliable)
results = model.track(source=0, show=True)
prices = {
    "Blue_doughnut": 1.50,
    "Bread_with_chocolate_filling": 2.50,
    "Bread_with_filling": 2.00,
    "Bread_with_milk_filling": 2.20,
    "Bread_with_milk_tea_filling": 2.30,
    "Bread_with_pandan_filling": 2.40,
    "Green_doughnut": 1.50,
    "Soft_bread": 2.00,
    "Soft_sesame_bun": 2.50,
    "ball": 1.00,
    "bean": 1.10,
    "castera": 2.00,
    "cookie": 1.50,
    "cream": 2.50,
    "foot": 1.20,
    "guabaegi": 1.30,
    "heart": 1.40,
    "soboro": 1.50
}

cart = {}
for result in results:
    for box in result.boxes:
        class_id = int(box.cls)
        class_name = result.names[class_id]
        if class_name in cart:
            cart[class_name] += 1
        else:
            cart[class_name] = 1

#calculate total cost
total_cost = 0.0
for item, quantity in cart.items():
    total_cost += prices[item] * quantity
print("Cart items:")
for item, quantity in cart.items():
    print(f"{item}: {quantity} x ${prices[item]:.2f} = ${prices[item] * quantity:.2f}")
print(f"Total cost is : ${total_cost:.2f}")
```

```python
# Save receipt to a .txt file with current date and time as filename
def save_receipt(cart, prices, total_cost):
    from datetime import datetime
    filename = datetime.now().strftime("receipt_%Y-%m-%d_%H-%M-%S.txt")
    with open(filename, "w") as f:
        f.write("Cart items:\n")
        for item, quantity in cart.items():
            line = f"{item}: {quantity} x ${prices[item]:.2f} = ${prices[item] * quantity:.2f}\n"
            f.write(line)
        f.write(f"\nTotal cost is : ${total_cost:.2f}\n")
    print(f"Receipt saved as '{filename}'")

# Call the function after printing the receipt
save_receipt(cart, prices, total_cost)
```
